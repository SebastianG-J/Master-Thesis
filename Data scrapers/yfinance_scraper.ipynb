```python
# Define the ticker for S&P 500 Index
ticker = "^SPX"
start_date = "2022-01-01"
end_date = "2025-01-01"
interval = "1d"

# Download historical data
spx_yf = yf.download(ticker,
                                  start_date,
                                  end_date,
                                  interval = interval)

# Display the first few rows
display(spx_yf.head())
display(spx_yf.tail())
```

Using the SPY tracker with VIX index

```python
import yfinance as yf
import pandas as pd

# Define tickers
price_ticker = "SPY"   # tracks natural aggregation of all constituent companies
vix_ticker = "^VIX"     # Volatility Index

# Define time range and interval
start_date = "2022-01-01"
end_date = "2025-01-01"
interval = "1d"

# Download data

spx_data = yf.download(price_ticker, start=start_date, end=end_date, interval=interval, auto_adjust=True)
vix_data = yf.download(vix_ticker, start=start_date, end=end_date, interval=interval)[['Close']]
vix_data.rename(columns={'Close': 'VIX_Close'}, inplace=True)

# Merge VIX with SPX data
combined_data = spx_data.merge(vix_data, left_index=True, right_index=True, how='left')

display(combined_data.head())

# Flatten MultiIndex columns
combined_data.columns = ['_'.join(col).strip() if isinstance(col, tuple) else col for col in combined_data.columns]

df_price_vix = combined_data.rename(columns={
    'Open_SPY': 'Open',
    'High_SPY': 'High',
    'Low_SPY': 'Low',
    'Close_SPY': 'Close',
    'Volume_SPY': 'Volume',
    'VIX_Close_^VIX': 'VIX'

})


display(df_price_vix.head())

save_path = "/content/drive/MyDrive/Masters Thesis/Colab notebook/Data scrapers/spy_vix_combined.csv"
df_price_vix.to_csv(save_path)
```
