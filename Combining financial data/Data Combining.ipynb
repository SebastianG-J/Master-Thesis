```python
# Get unique values from 'Author_Handle' column
unique_values = df_filtered['Author_Handle'].dropna().unique()

# Convert the unique values into a list
unique_values_list = list(unique_values)

# Define a function to split the list into chunks of 15 items per line
def chunks(lst, n):
    for i in range(0, len(lst), n):
        yield lst[i:i + n]

# Create the formatted output as a single list with 15 items per line
formatted_output = []
for chunk in chunks(unique_values_list, 15):
    formatted_output.append(", ".join([repr(item) for item in chunk]))  # Format each chunk

# Join everything together with commas between the chunks
final_output = "[" + ",\n ".join(formatted_output) + "]"

# Print the entire list as one continuous list with 15 items per row
print(final_output)
```

```python
# Get unique values from 'Author_Handle' column
unique_values = df_filtered['Author_Handle'].dropna().unique()

# Convert the unique values into a list
unique_values_list = list(unique_values)

# Function to split the list into chunks of 100 items
def chunks(lst, n):
    for i in range(0, len(lst), n):
        yield lst[i:i + n]

# Initialize a list to hold the formatted lists
formatted_output = []

# Iterate over the chunks of 100 items
for chunk in chunks(unique_values_list, 100):
    # Format the chunk with 15 items per line
    formatted_chunk = ",\n ".join([repr(item) for item in chunk])
    # Create a properly formatted list with 100 items
    formatted_output.append(f"[{formatted_chunk}]")

# Join all lists with two newlines between them
final_output = "\n\n".join(formatted_output)

# Print the final output to see the result
print(formatted_output)
```

## Combine the follower count and data

```python
# Load in follower data
followers = pd.read_excel("/content/drive/My Drive/Masters Thesis/Colab notebook/X follower count data/x_follower_counts_merged_cleaned.xlsx")

# renaming column to match in both sheets
followers.rename(columns={"Username": "Author_Handle"}, inplace=True)

# make as a dictionary
follower_dict = followers.set_index("Author_Handle")["Follower Count"].to_dict()

# Update follower count in df1 if Author_Handle exists in df2
df_filtered["Follower_Count"] = df_filtered["Author_Handle"].map(follower_dict)
```
